```python

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score


X , label = make_blobs(n_samples = 100 , centers = 3 , n_features = 2 , cluster_std =0.2,random_state = 42)

print(X)
print("label",label)

df = pd.DataFrame(X, columns = ["x" , "y"])
print(df["x"])
print("y",df["y"])

# ax1 = df.plot.scatter(x='length',y='width',
#                    c='species',)
plt.plot(df["x"],df["y"],marker="o",markeredgecolor="blue",markerfacecolor="black")
plt.show()

from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances ,  silhouette_score
kmeans = KMeans(n_clusters = 3)
pred = kmeans.fit_predict(df) 
plt.scatter(df["x"], df["y"], c='r')

plt.show()
print(kmeans.cluster_centers_)
print("label",kmeans.labels_) 
# display label of evry data points

score = silhouette_score(df, kmeans.labels_, metric='euclidean')
print("silhouette_score: ",score)

print("Accuracy:",metrics.accuracy_score(kmeans.labels_ , label))
precision = precision_score(pred, kmeans.labels_ , average=None)
recall = recall_score(pred, kmeans.labels_,average=None)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))

# Accuracy: 0.34
# precision: [1. 1. 1.]
# recall: [1. 1. 1.]
```
